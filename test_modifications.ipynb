```python
from scipy.stats import chi2
from scipy.stats import norm

import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt

import autograd.numpy as np

import confint_withoutg_linear_modification as confint_withoutg_linear_modification
```

```python
np.random.seed(0)
n = 100
d = 1
alpha = 0.1
dist = "gaussian"
df = 3
```

```python
# Create problem data
if (dist == "gaussian"):
    X = np.random.randn(n)

elif (dist == "gamma"):
    X = np.random.gamma(shape=1.0, size=n)

elif (dist == "chisq"):
    X = np.random.chisquare(df=df, size=n)

elif(dist == "uniform"):
    X = np.random.uniform(low=-3, high=3, size=n)

elif(dist == "mixture"):
    n_minus1 = n // 2
    n_plus1 = n - n_minus1
    X = np.concatenate([-2 + np.random.randn(n_minus1), 2 + np.random.randn(n_plus1)])
else:
    print("ERROR: unsupported distribution")

X = np.sort(X)
```

```python
thread_num = 8
M = 12.0
tau_max = 1e5
mu = 8.0
solver_choice = "MOSEK"
```

```python
conf_int = confint_withoutg_linear_modification.confint(n, X, alpha)
```

```python
conf_int.compute_pw_conf_ints(thread_num=thread_num, modification=0, M=M, tau_max=tau_max, mu=mu)
conf_int.remove_nan()
lo_0 = np.array(conf_int.lo)
hi_0 = np.array(conf_int.hi)
opt_pts_0 = np.array(X[conf_int.idxes_of_design_pts_to_opt])

conf_int.compute_pw_conf_ints(thread_num=thread_num, modification=1, M=M, tau_max=tau_max, mu=mu)
conf_int.remove_nan()
lo_1 = np.array(conf_int.lo)
hi_1 = np.array(conf_int.hi)
opt_pts_1 = np.array(X[conf_int.idxes_of_design_pts_to_opt])

conf_int.compute_pw_conf_ints(thread_num=thread_num, modification=2, M=M, tau_max=tau_max, mu=mu)
lo_2 = np.array(conf_int.lo)
hi_2 = np.array(conf_int.hi)
opt_pts_2 = np.array(X[conf_int.idxes_of_design_pts_to_opt])


conf_int.compute_pw_conf_ints(thread_num=thread_num, modification=3, M=M, tau_max=tau_max, mu=mu)
lo_3 = np.array(conf_int.lo)
hi_3 = np.array(conf_int.hi)
opt_pts_3 = np.array(X[conf_int.idxes_of_design_pts_to_opt])
```

```python
plt.figure()
# plt.subplot(221)
plt.plot(opt_pts_0, lo_0, '-^', linewidth=2.0, label="mod 0", color="blue")
plt.plot(opt_pts_0, hi_0, '-^', linewidth=2.0,  color="blue")
plt.plot(opt_pts_0, norm.pdf(opt_pts_0), label="ground truth")

# plt.subplot(222)
plt.plot(opt_pts_1, (lo_1), '-.', linewidth=2.0, label="mod 1", color="red")
plt.plot(opt_pts_1, (hi_1), '-.', linewidth=2.0, color="red")

# plt.subplot(223)
plt.plot(opt_pts_2, (lo_2), '--o', linewidth=2.0, label="mod 2", color="black")
plt.plot(opt_pts_2, (hi_2), '--o', linewidth=2.0, color="black")

# plt.subplot(224)
plt.plot(opt_pts_3, (lo_3), ':.', linewidth=2.0, label="mod 3", color="green")
plt.plot(opt_pts_3, (hi_3), ':.', linewidth=2.0, color="green")
plt.legend(loc=2)
plt.grid()
```
