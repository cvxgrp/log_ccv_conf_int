```python
from scipy.stats import chi2
from scipy.stats import norm

import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt

import autograd.numpy as np

import confint_withoutg_linear_modification as confint_withoutg_linear_modification
```

```python
np.random.seed(0)
n = 100
d = 1
alpha = 0.05
dist = "gaussian"
df = 3
```

```python
# Create problem data
if (dist == "gaussian"):
    X = np.random.randn(n)

elif (dist == "gamma"):
    X = np.random.gamma(shape=1.0, size=n)

elif (dist == "chisq"):
    X = np.random.chisquare(df=df, size=n)

elif(dist == "uniform"):
    X = np.random.uniform(low=-3, high=3, size=n)

elif(dist == "mixture"):
    n_minus1 = n // 2
    n_plus1 = n - n_minus1
    X = np.concatenate([-2 + np.random.randn(n_minus1), 2 + np.random.randn(n_plus1)])
else:
    print("ERROR: unsupported distribution")

X = np.sort(X)
```

```python
thread_num = 8
M = 8.0
tau_max = 1e3
mu = 8.0
max_iters = 20
solver_choice = "MOSEK"
```

```python
conf_int = confint_withoutg_linear_modification.confint(n, X, alpha)
opt_pts = X[conf_int.idxes_of_design_pts]
```

```python
conf_int.compute_pw_conf_ints(thread_num, M, tau_max, mu, max_iters, modification=3)
```

```python
plt.figure()
plt.subplot(111)
plt.plot(opt_pts, np.log(conf_int.lo), '-o', linewidth=1.0, label="L")
plt.plot(opt_pts, np.log(conf_int.hi), '-o', linewidth=1.0, label="H")
plt.plot(opt_pts, np.log(norm.pdf(opt_pts)), label="ground truth")
plt.legend(loc=4)
plt.ylabel("$\log f$")
```
